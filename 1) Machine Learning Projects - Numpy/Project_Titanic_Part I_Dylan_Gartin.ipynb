#### 4.5) b) Fill missing values of 'Age' field with the median age of the passenger class and sex that you found for the question above. Use the apply method with lambda function. 


```python
# use median age to fill in the above
df_all['Age'] = df_all.groupby(['Pclass', 'Sex'])['Age'].apply(lambda x: x.fillna(x.median()))
```

#### 4.5) c) Fill missing values of 'Fare' field with the median fare of the passenger class and embarked location that you found for the question above. Use the apply method with lambda function. 

```python
#CHECK THIS ONE
df_all['Fare'] = df_all.groupby(['Pclass', 'Embarked'])['Fare'].apply(lambda x: x.fillna(x.median()))
```

#### 4.5) d) Fill missing values of 'Cabin' field with the 'NA' value.

```python
# NA
df_all['Cabin'] = df_all['Cabin'].apply(lambda s: s[0] if pd.notnull(s) else 'NA')
```

```python
# run this to see which value has most then use that as the fill data
df_all['Embarked'].value_counts().idxmax()
```

```python
df_all.loc[df_all['Embarked'].isnull(), 'Embarked'] = 'S'
```
